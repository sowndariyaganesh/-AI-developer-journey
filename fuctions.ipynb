```python
contacts=[]
```

```python
def add_contact():

    print("\n--- Add New Contact ---")

    name=input("Enter Name: ").strip().title()
    phone=input("Enter Phone: ").strip()
    email=input("Enter Email: ").strip().lower()

    contact={
    "name":name,
    "phone":phone,
    "email":email
}
    contacts.append(contact)
    print("contact added")
```

```python
add_contact()
```

```python
def view_all():
  print("\n---All Contacts---")
  if len(contacts)==0:
    print("No Contacts Found")
    return
  for i,c in enumerate(contacts,1):
    print(f"{i}. {c['name']}"
              f"   Phone: {c['phone']}"
              f"   Email: {c['email']}")
```

```python
def search_contact():
    print("\n--- Search Contact ---")
    query = input("Enter name to search: ").strip().lower()
    found = []

    for c in contacts:
        if query in c["name"].lower():
            found.append(c)

    if found:
        print(f"Found {len(found)} result(s):")
        for c in found:
            print(f"  {c['name']} | {c['phone']} | {c['email']}")
    else:
        print(f"No contact found with name '{query}'")

```

```python
def delete_contact():
    print("\n--- Delete Contact ---")
    name = input("Enter exact name to delete: ").strip().lower()

    for c in contacts:
        if c["name"].lower() == name:
            contacts.remove(c)
            print(f"Deleted: {c['name']}")
            return

    print(f"No contact named '{name}' found.")
```

```python
def show_menu():
    print("\n========== Contact Book ==========")
    print(f"  Contacts stored: {len(contacts)}")
    print("  1. Add contact")
    print("  2. View all contacts")
    print("  3. Search contact")
    print("  4. Delete contact")
    print("  5. Quit")
    print("==================================")


def main():
    print("Welcome to your Contact Book!")

    while True:
        show_menu()
        choice = input("Choose (1-5): ").strip()

        if   choice == "1": add_contact()
        elif choice == "2": view_all()
        elif choice == "3": search_contact()
        elif choice == "4": delete_contact()
        elif choice == "5":
            print("Bye! Day 2 complete.")
            break
        else:
            print("Invalid choice. Enter 1-5.")


main()   # start the program
```
