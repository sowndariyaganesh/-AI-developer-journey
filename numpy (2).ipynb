```python
import numpy as np

# Dataset: 10 students × 5 subjects
# Columns: Maths, Physics, Chemistry, English, CompSci
scores = np.array([
    [88, 92, 79, 85, 95],
    [74, 68, 82, 91, 77],
    [56, 61, 58, 70, 63],
    [93, 88, 94, 87, 96],
    [45, 52, 49, 61, 55],
    [81, 79, 85, 76, 88],
    [67, 72, 69, 74, 71],
    [91, 86, 90, 83, 92],
    [62, 58, 64, 69, 60],
    [78, 84, 76, 80, 82],
])
subjects = ["Maths", "Physics", "Chemistry", "English", "CompSci"]

# --- Per-student analysis (axis=1) ---
student_avg = np.mean(scores, axis=1)
student_best = np.max(scores, axis=1)

print("--- Student Report ---")
for i in range(10):
    g = ("A" if student_avg[i] >= 85 else
         "B" if student_avg[i] >= 70 else
         "C" if student_avg[i] >= 55 else "F")
    print(f"S{i+1}: avg={student_avg[i]:.1f}  best={student_best[i]}  grade={g}")

# --- Rankings ---
ranked = np.argsort(student_avg)[::-1]   # [::-1] reverses to descending
print("\n--- Rankings ---")
for rank, idx in enumerate(ranked, 1):
    print(f"  #{rank}: Student {idx+1}  avg={student_avg[idx]:.1f}")

# --- Per-subject analysis (axis=0) ---
subject_avg = np.mean(scores, axis=0)
hardest = np.argmin(subject_avg)
print(f"\n--- Subject Averages ---")
for i, s in enumerate(subjects):
    print(f"  {s:12} avg={subject_avg[i]:.1f}")
print(f"Hardest subject: {subjects[hardest]}")

# --- Boolean indexing: who failed? who topped? ---
failing = np.where(student_avg < 60)[0] + 1
toppers = np.where(student_avg >= 85)[0] + 1
print(f"\nStudents needing help: {list(failing)}")
print(f"Top performers:        {list(toppers)}")

# --- Normalise all scores to 0-1 ---
normalised = (scores - scores.min()) / (scores.max() - scores.min())
print(f"\nNormalised — min: {normalised.min():.2f}, max: {normalised.max():.2f}")
```
