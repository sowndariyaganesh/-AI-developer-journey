```python
data['burnout_level'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['burnout_level']=le.fit_transform(data['burnout_level'])
data
```

```python
from sklearn.impute import KNNImputer
knn=KNNImputer(n_neighbors=10)
data['burnout_level']=knn.fit_transform(data[['burnout_level']])
data['burnout_level']
```

```python
data['burnout_level'].isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data['experience_years']=sc.fit_transform(data[['experience_years']])
data['daily_work_hours']=sc.fit_transform(data[['daily_work_hours']])
data['sleep_hours']=sc.fit_transform(data[['sleep_hours']])
data['caffeine_intake']=sc.fit_transform(data[['caffeine_intake']])
data['bugs_per_day']=sc.fit_transform(data[['bugs_per_day']])
data['commits_per_day']=sc.fit_transform(data[['commits_per_day']])
data['meetings_per_day']=sc.fit_transform(data[['meetings_per_day']])
data['screen_time']=sc.fit_transform(data[['screen_time']])
data['exercise_hours']=sc.fit_transform(data[['exercise_hours']])
data['stress_level']=sc.fit_transform(data[['stress_level']])

data
```

```python
data.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
knn=KNNImputer(n_neighbors=5)
data['stress_level']=knn.fit_transform(data[['stress_level']])
data
```
